```python
import pandas as pd
from pykrx import stock
import tqdm


iem = pd.read_csv('iem_info.csv')
```

```python
iem['code'] = iem['iem_cd'].apply(lambda x:x[1:])
```

```python
data_code = iem['code'].tolist()
```

```python
data = stock.get_market_ohlcv_by_date("20160101", "20210801", data_code[0])
data = data.T.reset_index(drop=True)
data.insert(0,'코드',data_code[:1]*5)
data
```

```python
null_code = []
for code in tqdm.tqdm(data_code[1:]):
    try:
        tmp = stock.get_market_ohlcv_by_date("20160101", "20210801", code)
        tmp = tmp.T.reset_index(drop=True)
        tmp.insert(0,'코드',[code]*5)
        data = pd.concat([data,tmp])
    except:
        null_code.append(code)
```
