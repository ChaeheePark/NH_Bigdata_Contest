```python
# 주가 정보 전처리

# 컬럼을 기존 데이터에 맞게 변경
stock_col = list(stock.columns)
def get_d(day):
    return int(re.sub('[^0-9]+','',day[:10]))
stock_col[1:] = list(map(get_d,stock_col[1:]))
stock.columns = stock_col

# 코드가 int로 바뀌면서 앞의 0이 사라지는 경우
stock['코드'] = stock['코드'].astype('str')
def change(text):
    if len(text)<6:
        tmp = 6-len(text)
        return ('0'*tmp)+text
    else:
        return text

stock['코드'] = stock['코드'].apply(lambda x: change(x))
```


### 모델 학습: LGBMRegressor 사용

```python
submission.loc[test_data.index,'hold_d'] = list(range(1,70597))
```

```python
for train_data, train_label, test_data in zip(train_data_model, train_label_model, test_data_model):
    models = []
    train_data.reset_index(drop=True,inplace=True)
    train_label.reset_index(drop=True,inplace=True)
    
    folds = KFold(n_splits=10)
    for train_idx, val_idx in tqdm(folds.split(train_data),total=folds.get_n_splits(), desc="k-fold"):

        train_x = train_data.iloc[train_idx, :]
        train_y = train_label[train_idx]
        val_x = train_data.iloc[val_idx, :]
        val_y = train_label[val_idx]

        model = LGBMRegressor(objective= "regression",
                              max_depth= 5,
                              n_estimators= 3000,
                              learning_rate= 0.01,
                              num_leaves = 30)

        model.fit(train_x, train_y,
                  eval_set=[(val_x, val_y)],
                  eval_metric=["rmse"],
                  early_stopping_rounds=300,
                  verbose=500)

        models.append(model)
        
    result = []
    for i in models:
        result.append(i.predict(test_data))
    predict = np.mean(result, axis = 0)
    
    submission.loc[test_data.index,'hold_d'] = predict
```
