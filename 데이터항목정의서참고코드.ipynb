```python
# 결측치 제거

cus_info.drop(cus_info[cus_info['tco_cus_grd_cd']==99].index,axis=0,inplace=True)
```

```python
#bse_dt 업데이트 기준날짜마다 count해줌 -> 임의가 아닌, 고객의 매수 매도 날짜임 (증거, 카카오톡 사진 첨부)

temp = stk_bnc_hist.groupby('act_id')['bse_dt'].count().reset_index()
temp.rename(columns={'bse_dt':'buy_sell_cnt'},inplace=True)

cus_info = pd.merge(cus_info,temp,on='act_id',how='left')
cus_info.dropna(inplace=True)

cus_info['buy_sell_cnt'] = cus_info['buy_sell_cnt'].astype(int)
```

```python
# 고객 자산 구간 선택 / 고객등급은 버림 왜냐면 buy_sell cnt로 세기 때문에 많이 사고 팔았음 cnt가 높을거니까, 고객자산구간이 우리에게 더 적합한 데이터라고 판단.

# 고객자산구간 별, 매도 매수 횟수로 공격형/ 중립형/ 방어형 구분

quant_save = pd.DataFrame()

quant_save['1/3'] = cus_info.groupby('cus_aet_stn_cd')['buy_sell_cnt'].quantile(0.3)
quant_save['2/3'] = cus_info.groupby('cus_aet_stn_cd')['buy_sell_cnt'].quantile(0.6)
quant_save['3/3'] = cus_info.groupby('cus_aet_stn_cd')['buy_sell_cnt'].quantile(0.9)

quant_save

# 3/3 quantile(1)로 했을때 값 차이가 너무 심함, 이상치 인지 아닌지 확인 필요
```

```python
# 그래서 max변수 만들어서 이상치?는 아니지만 초 단타 쟁이들 분류해줌

quant_save['max'] = cus_info.groupby('cus_aet_stn_cd')['buy_sell_cnt'].quantile(1)
```

```python
# 1 : 방어형, 2 : 중립형, 3 : 공격형, 4: 초단타(아마 봇)

#투자유형 생성
cus_info['inv_type'] = cus_info.apply(lambda x : 1 if x['buy_sell_cnt'] < quant_save.loc[x['cus_aet_stn_cd']]['1/3'] else (2 if x['buy_sell_cnt'] < quant_save.loc[x['cus_aet_stn_cd']]['2/3'] else (3 if x['buy_sell_cnt'] < quant_save.loc[x['cus_aet_stn_cd']]['3/3'] else 4)),axis=1)

# hist_percentage 생성위해 삭제 보류
# cus_info.drop(['buy_sell_cnt'],axis=1,inplace=True)
```

```python
temp = pd.to_datetime(stk_bnc_hist.groupby('act_id')['bse_dt'].max(),format='%Y%m%d') - pd.to_datetime(stk_bnc_hist.groupby('act_id')['bse_dt'].min(),format='%Y%m%d')

temp = pd.DataFrame(temp.map(lambda x : x.days)).reset_index()
temp.rename(columns={'bse_dt':'total_date'},inplace=True)
cus_info = pd.merge(cus_info,temp,how='left')
```
