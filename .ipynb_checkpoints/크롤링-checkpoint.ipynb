```python
import pandas as pd

iem=pd.read_csv('data/iem_info_20210902.csv')
```

```python
iem_tp=iem.loc[:1078]
```

```python
from tqdm import tqdm
from selenium import webdriver
import time
```

```python

chromedriver = 'C:/Users/chaeh/chromedriver.exe'
driver = webdriver.Chrome(chromedriver)


for i in tqdm(range(len(iem_tp))):
   
    driver.get('https://datalab.naver.com/keyword/trendSearch.naver')
    name=iem_tp['iem_krl_nm'][i]
    search_iem = driver.find_element_by_id("item_keyword1")
    search_iem.send_keys(name)
    driver.find_element_by_xpath('//*[@id="startYear"]').click()
    driver.find_element_by_xpath('//*[@id="startYearDiv"]/ul/li[1]/a').click()
    driver.find_element_by_xpath('//*[@id="startMonth"]').click()
    driver.find_element_by_xpath('//*[@id="startMonthDiv"]/ul/li[1]/a').click()
    driver.find_element_by_xpath('//*[@id="startDay"]').click()
    driver.find_element_by_xpath('//*[@id="startDayDiv"]/ul/li[1]/a').click()

    driver.find_element_by_xpath('//*[@id="endMonth"]').click()
    driver.find_element_by_xpath('//*[@id="endMonthDiv"]/ul/li[1]/a').click()
    driver.find_element_by_xpath('//*[@id="endDay"]').click()
    driver.find_element_by_xpath('//*[@id="endDayDiv"]/ul/li[1]/a').click()

    driver.find_element_by_xpath('//*[@id="content"]/div/div[2]/div[1]/div/form/fieldset/a').click()
    
    time.sleep(2)
    driver.find_element_by_xpath('//*[@id="content"]/div/div[1]/div[1]/div/div/div/div/div/div[1]/div[4]/a').click()
    #driver.close()
```
