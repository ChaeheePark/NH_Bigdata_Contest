```python
import pandas as pd
```

```python
cus_info = pd.read_csv('cus_info.csv')
stk_bnc_hist = pd.read_csv('stk_bnc_hist.csv')
iem_info = pd.read_csv('iem_info_20210902.csv')
stk_hld_train = pd.read_csv('stk_hld_train.csv')
stk_hld_test = pd.read_csv('stk_hld_test.csv')
```

# 0915 작업

```python
# 결측치 제거

cus_info.drop(cus_info[cus_info['tco_cus_grd_cd']==99].index,axis=0,inplace=True)
```

```python
pd.options.display.float_format = '{:.3f}'.format
cus_info.corr()['tco_cus_grd_cd']

#고객등급, 자산구간 높은 상관관계
```

```python
#bse_dt 업데이트 기준날짜마다 count해줌 -> 임의가 아닌, 고객의 매수 매도 날짜임 (증거, 카카오톡 사진 첨부)

temp = stk_bnc_hist.groupby('act_id')['bse_dt'].count().reset_index()
temp.rename(columns={'bse_dt':'buy_sell_cnt'},inplace=True)
temp.head(5)
```

```python
cus_info = pd.merge(cus_info,temp,on='act_id',how='left')
cus_info.head(5)
```

```python
len(cus_info[cus_info['buy_sell_cnt'].isna()])
```

```python
cus_info.dropna(inplace=True)
```

```python
cus_info['buy_sell_cnt'] = cus_info['buy_sell_cnt'].astype(int)
```

```python
# 고객 자산 구간 선택 / 고객등급은 버림 왜냐면 buy_sell cnt로 세기 때문에 많이 사고 팔았음 cnt가 높을거니까, 고객자산구간이 우리에게 더 적합한 데이터라고 판단.

# 고객자산구간 별, 매도 매수 횟수로 공격형/ 중립형/ 방어형 구분

quant_save = pd.DataFrame()

quant_save['1/3'] = cus_info.groupby('cus_aet_stn_cd')['buy_sell_cnt'].quantile(0.3)
quant_save['2/3'] = cus_info.groupby('cus_aet_stn_cd')['buy_sell_cnt'].quantile(0.6)
quant_save['3/3'] = cus_info.groupby('cus_aet_stn_cd')['buy_sell_cnt'].quantile(0.9)

quant_save

# 3/3 quantile(1)로 했을때 값 차이가 너무 심함, 이상치 인지 아닌지 확인 필요,
```

```python
quant_save['max'] = cus_info.groupby('cus_aet_stn_cd')['buy_sell_cnt'].quantile(1)

quant_save
# 그래서 max변수 만들어서 이상치?는 아니지만 초 단타 쟁이들 분류해줌
```

```python
# 1 : 방어형, 2 : 중립형, 3 : 공격형, 4: 초단타(아마 봇)

#투자유형 생성
cus_info['inv_type'] = cus_info.apply(lambda x : 1 if x['buy_sell_cnt'] < quant_save.loc[x['cus_aet_stn_cd']]['1/3'] else (2 if x['buy_sell_cnt'] < quant_save.loc[x['cus_aet_stn_cd']]['2/3'] else (3 if x['buy_sell_cnt'] < quant_save.loc[x['cus_aet_stn_cd']]['3/3'] else 4)),axis=1)

# hist_percentage 생성위해 삭제 보류
# cus_info.drop(['buy_sell_cnt'],axis=1,inplace=True)

cus_info
```

```python
#분류된 투자 성향별 count

cus_info['inv_type'].value_counts()
```

```python
cus_info.corr()
```

```python
# 상관관계 있는 features

# 연령구간 & LiftStage
# 고객자산구간 & 고객등급 

# 고객등급, LifeStage 삭제처리 
# 고객등급은 새로 만든 inv_type으로 대체 가능하다 판단. 고객자산구간은 대체 불가

# life stage는 너무 TMi (over fitting 위험) 라고 판단, 연령구간만 사용해도 충분하다고 판단.

cus_info.drop(['tco_cus_grd_cd','lsg_sgm_cd'],axis=1,inplace=True)
cus_info
```

```python
# 국내주식 외 거래 삭제 + 주거래상품군 feature 삭제 처리

cus_info = cus_info[cus_info['mrz_pdt_tp_sgm_cd']==2]
cus_info.drop(['mrz_pdt_tp_sgm_cd'],axis=1,inplace=True)
cus_info
```

```python
# cus_info 안 주거래업종(mrz_btp_dit_cd)은 비매매, 매매 이렇게 두가지로만 나누기로함(이유는 x년도 12월기준 한달 데이터만 파악하기 때문에, 부족한 데이터라고 판단. )
# 대신 실제 고객의 거래 내역을 보고 주거래업종을 파악해 보기로 함 -> 채희 담당

cus_info['not_buy'] = cus_info.apply(lambda x : 1 if x['mrz_btp_dit_cd']==16 else 0,axis=1)
cus_info.drop(['mrz_btp_dit_cd'],axis=1,inplace=True)

cus_info
```

```python
# 분위별 값 구하기

import numpy as np

temp = []
temp_cnt = np.arange(0,1,0.1)


for i in temp_cnt:
    temp.append(cus_info['buy_sell_cnt'].quantile(i))
    
temp
```

```python
temp_list = []


for i in cus_info['buy_sell_cnt']:
    find = 0
    for cnt, j in enumerate(temp):
        if i<=j:
            temp_list.append(cnt)
            find=1
            break
    if find==0:
        temp_list.append(10)
```

```python
# hold_d에 바로 곱해주기 위해 0~1사이 값으로 만들어줌

temp_list = [int(i)/10 for i in temp_list]

len(temp_list)
```

```python
# hist_d에 곱해줄 값

cus_info['buy_sell_quant'] = temp_list

cus_info
```

```python
# over fitting 우려로 삭제.

cus_info.drop(['buy_sell_cnt'],axis=1,inplace=True)
cus_info
```
